##  Esercizio 1 - Oscillatore Forzato

Considerando l'equazione per il moto di una molla sottoposta ad una forza esterna variabile $F(t)$:

$$
m \frac{d^2x}{dt^2} = -kx -C  \frac{dx}{dt} + \frac{F(t)}{m};
$$

dove $k$ è la costante eleastica della molla e $C$ la costante legata alla dissipazione interna della molla.

Definendo:  
$$
\gamma = \frac{C}{2m}
$$

e 
$$
\omega_0^2 = \frac{k}{m}
$$

possiamo riformulare l'equazione differenziale come:
$$
\frac{d^2x}{dt^2} +2\gamma \frac{dx}{dt} + \omega_0^2 x - \frac{F(t)}{m}= 0.
$$

$$

C  = 0.5 \; N s/m\\
$$

e dalle condizioni inziali:
$$
x_0 = 0;
v_0 = 0;
$$

produrre uno script python che:
1. risolva l'equazione differenziale per l'oscillatore forzato in modo tale che la funzione che descrive la  forza esterna $F(t)$ possa essere passata come parametro;
2. produca il grafico di $F(t)$, $x(t)$ e $v(t)$ per una forza sinusoidale $F(t) = 2 \sin \omega_f t$ con un  valore di $\omega_f$ a scelta;
3. produca il grafico dell'ampiezza di oscillazione massima in funzione del valore di $\omega_f$ per la forza sinusoidale $F(t) = 2 \sin \omega_f t$;

L'equazione del moto del pendolo semplice con massa $m$ e lunghezza del filo $l$ :

$$
ml \frac{d^2 \theta}{d t^2} = -mg \sin(\theta);
$$


$$
 \frac{d^2 \theta}{d t^2} = -\frac{g}{l} \sin(\theta)
$$


Definendo:
$$
\frac{d \theta}{d t} = \omega
$$

Il moto del pendolo può essere descritto dalle due equazioni:

$$
\begin{cases}
\begin{eqnarray}
\frac{d \theta}{d t} &  = & \omega \\
 \frac{d \omega}{d t}&  =  & -\frac{g}{l} \sin(\theta) \\
\end{eqnarray}
\end{cases}
$$

Pe risolvere le equazioni del moto numericamente combiniamo le variabili $\theta$ e $\omega$ nel vettore ${\bf r}(\theta, \omega)$.

    2. $l = 0.5 \;m$,  $\theta_0 = 30^{\circ}$
